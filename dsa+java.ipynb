### INTRODUCTION
A Public Transportation Routing System helps passengers find the best and most efficient routes using buses, trains, or other public transport options. To achieve this, the system uses:

**Priority Queue for Route Optimization**:
The priority queue is used to compute the best routes by ordering the alternatives according to travel time, stops, or congestion levels. It automatically determines the shortest or fastest route for the sake of overall travel effectiveness.

**Linked List for Transportation Schedule Tracking** –A linked list data structure helps manage transportation timetables by maintaining an ordered and readily updatable list of arrival and departure times. This allows for timely updating when timetables are changed, delivering accurate and real-time tracking of public transportation services.

**Efficient Route Planning** – Using effective data structures like priority queues and linked lists, the system efficiently computes and suggests the most efficient available routes. This enhances public transport reliability through reduced delays and enhanced passenger satisfaction.

```python
import heapq


class Station:
    def __init__(self, station_id, distance=float('inf')):
        self.station_id = station_id  
        self.distance_from_start = distance  
        self.adjacency_list = []  # this list is used to hold the values of references to other station.
        self.previous_station = None  #it is used to keep track of previous station.

class MinHeap: # Priority queue
    def __init__(self):
        self.elements = []  # to store the elements of minheap DS.

    def push(self, priority, station):
        heapq.heappush(self.elements, (priority, station))
 
    def pop(self): #removes the elemment with highest priority 
        return self.elements.pop(0) if self.elements else None # if it contains elemt it returns the first elemnt, if no elemnmt then return 0.

    def is_empty(self):
        return len(self.elements) == 0


class ScheduleEntry: 
    def __init__(self, route_id, start_time, end_time): #this contains information about route and timing. 
        self.route_id = route_id  
        self.start_time = start_time  
        self.end_time = end_time  
        self.next_entry = None  # the nextentry attribute links schedulentry to next entry in linked list.

class ScheduleList: # this class manages the linkedlist of schedule entry.
    def __init__(self): # initializes a new instance of the class.
        self.head_entry = None  

    def add_schedule(self, route_id, start_time, end_time): # from this we can add a new entry to linked list 
        new_entry = ScheduleEntry(route_id, start_time, end_time)
        if not self.head_entry: # checks whether the list in empty or not
            self.head_entry = new_entry # first entry to the list 
        else:
            current = self.head_entry # if the list in not empty the current variable points to the first entry  in the list
            while current.next_entry: # this loop iterates through the linked list until if find the last entry
                current = current.next_entry # it waits until it reaches the lst entry/
            current.next_entry = new_entry

    def get_schedules(self): # it will get all the schedule entries from linked list
        schedules = []
        current = self.head_entry # it points to first entry in linked list 
        while current:
            schedules.append((current.route_id, current.start_time, current.end_time))
            current = current.next_entry
        return schedules


class TransportationSystem: # this class is responsible for managing the stations and connections.
    def __init__(self):
        self.stations_graph = {}  # it is defined as dictionary becasue each station is contains a unique ID 
        self.schedule_list = ScheduleList()  # manages the schedules for the system
    def dijkstra(self, start_id):
        if start_id not in self.stations_graph: # checks whether start id is present in stations graph. if it not avaliable then it will retunr none 
            return None

        
        for station in self.stations_graph.values():# iterates over all the staions 
            station.distance_from_start = float('inf')# set to infinity it indicates that the intial distance from starting system is unknown
            station.previous_station = None # it will be used to construct the shortest path
        self.stations_graph[start_id].distance_from_start = 0 # starting positon 

        priority_queue = MinHeap()
        priority_queue.push(0, self.stations_graph[start_id])

        while not priority_queue.is_empty():
            current_station = priority_queue.pop()[1]

            for neighbor, travel_time in current_station.connections:
                new_distance = current_station.distance_from_start + travel_time
                if new_distance < neighbor.distance_from_start:
                    neighbor.distance_from_start = new_distance
                    neighbor.previous_station = current_station
                    priority_queue.push(new_distance, neighbor)

    def get_shortest_path(self, start_id, end_id):
        if start_id not in self.stations_graph or end_id not in self.stations_graph:
            return None

        path = []
        current_station = self.stations_graph[end_id]
        while current_station:
            path.append(current_station.station_id)
            current_station = current_station.previous_station
        return path[::-1] 

    def add_station(self, station_id, connections):
        if station_id not in self.stations_graph:
            self.stations_graph[station_id] = Station(station_id)
        for neighbor_id, travel_time in connections:
            if neighbor_id not in self.stations_graph:
                self.stations_graph[neighbor_id] = Station(neighbor_id)
            self.stations_graph[station_id].connections.append((self.stations_graph[neighbor_id], travel_time))

    def add_schedule(self, route_id, start_time, end_time):
        self.schedule_list.add_schedule(route_id, start_time, end_time)

    def get_schedules(self):
        return self.schedule_list.get_schedules()
```
